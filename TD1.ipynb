```python
import math
import numpy
import matplotlib.pyplot as plt
```

```python
def newtonRaphson(x0,e,N):
    print('\n\n*** NEWTON RAPHSON METHOD IMPLEMENTATION ***')
    x0 = float(x0)
    e = float(e)
    N = int(N)
    hist = []
    step = 1
    flag = 1
    condition = True
    while condition:
        if g(x0) == 0.0:
            print('Divide by zero error!')
            break
        
        x1 = x0 - f(x0)/g(x0)
        print('Iteration-%d, x1 = %0.6f, f(x1) = %0.6f and g(x1) = %0.6f' % (step, x1, f(x1), g(x1)))
        hist.append(x1)
        x0 = x1
        step = step + 1
        
        if step > N:
            flag = 0
            break
        
        condition = abs(f(x1)) > e
    
    if flag==1:
        print('\nRequired root is: %0.8f' % x1)
        plt.plot(hist)
    else:
        print('\nNot Convergent.')
        plt.plot(hist)
    
    
# Input Section
# x0 = input('Enter Guess: ')
e = 1e-6
N = 20
```

```python
def numerical_derivative(f,x, h=1e-5):
    """
    Create and return a function that represents the derivative of the input function.

    Parameters:
    - f: The function for which to compute the derivative. It should be a scalar function that
         takes a single numerical input and returns a numerical output.
    - h: The step size for numerical differentiation (default is 1e-5).

    Returns:
    - A function representing the derivative of the input function.
    """
    def derivative(x):
        return (f(x + h) - f(x - h)) / (2 * h)

    return derivative
```

# x -> sqrt(abs(x))

```python
def f(x):
    return math.sqrt(abs(x))

def g(x):
    return abs(x)/(2*math.sqrt(abs(x)))

newtonRaphson(0.1,e,N)
```

```python
def multi_newtonRaphson(x0,y0,e,N):
    print('\n\n*** NEWTON RAPHSON METHOD IMPLEMENTATION ***')
    x0 = float(x0)
    y0 = float(y0)
    e = float(e)
    N = int(N)
    hist = []
    step = 1
    flag = 1
    condition = True

    
    while condition:
        if g(x0) == 0.0:
            print('Divide by zero error!')
            break
        
        x1 = x0 - f(x0,y0)/gx(x0,y0)
        y1 = y0 - f(x1,y0)/gy(x1,y0)

        print('Iteration-%d, x1 = %0.6f, f(x1) = %0.6f and g(x1) = %0.6f' % (step, x1, f(x1,y0), gx(x1,y0)))
        print('Iteration-%d, y1 = %0.6f, f(x1) = %0.6f and g(x1) = %0.6f' % (step, y1, f(x1,y1), gy(x1,y1)))
        hist.append([x1,y1])
        x0 = x1
        y0 = y1
        step = step + 1
        
        if step > N:
            flag = 0
            break
        
        condition = abs(f(x1,y1)) > e
    
    if flag==1:
        print('\nRequired root is: %0.8f' % x1,y1)
        plt.plot(hist)
    else:
        print('\nNot Convergent.')
        plt.plot(hist)
    
```

# (x, y) ⟼ ln(1 + (x^2 + y^2 -1)^2)

```python
def f(x,y):
    return  math.log(1 + (x**2 + y**2 -1)**2)

def gx(x,y):
    return 2*x + y*0
def gy(x,y):
    return y**2 + y*0

multi_newtonRaphson(2,2,e,10000)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Pour le graphique 3D

def plot_function3d(f, x_range=(-5, 5), y_range=(-5, 5), num_points=400):
    """
    Tracer une fonction à deux variables.
    
    Paramètres:
    f (callable): La fonction à deux variables à tracer.
    x_range (tuple): La plage des valeurs de x à utiliser pour le tracé.
    y_range (tuple): La plage des valeurs de y à utiliser pour le tracé.
    num_points (int): Le nombre de points à utiliser pour le tracé.
    """
    x = np.linspace(x_range[0], x_range[1], num_points)
    y = np.linspace(y_range[0], y_range[1], num_points)
    x, y = np.meshgrid(x, y)
    z = f(x, y)
    
    fig = plt.figure(figsize=(12, 6))
    
    # Contour plot
    ax1 = fig.add_subplot(121)
    cp = ax1.contourf(x, y, z, cmap='viridis')
    fig.colorbar(cp, ax=ax1)
    ax1.set_title('Contour Plot')
    ax1.set_xlabel('x')
    ax1.set_ylabel('y')
    
    # 3D surface plot
    ax2 = fig.add_subplot(122, projection='3d')
    surf = ax2.plot_surface(x, y, z, cmap='viridis', antialiased=False)
    fig.colorbar(surf, ax=ax2)
    ax2.set_title('3D Surface Plot')
    ax2.set_xlabel('x')
    ax2.set_ylabel('y')
    ax2.set_zlabel('f(x, y)')
    
    plt.show()
```
